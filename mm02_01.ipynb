选择sbpjieshu，dbpjieshu，petco2jieshu，RRjieshu，spo2jieshu，HR10作为研究的分布（因为有较为规则的分布）

```python
dataB = pd.read_csv("data/B药生命体征.csv")
header_list = list(dataB.columns)
dataB = dataB.to_numpy(float).T
for data, title in zip(dataB, header_list):
    # 去掉nan
    data = data[~np.isnan(data)]
    # 绘制直方图
    sns.histplot(data, kde=True)
    plt.xlabel('value')
    plt.ylabel('frequency')
    plt.title(title)
    plt.show()
```

```python
dataR = pd.read_csv("data/R药生命体征.csv")
header_list = list(dataR.columns)
dataR = dataR.to_numpy(float).T
for data, title in zip(dataR, header_list):
    # 去掉nan
    data = data[~np.isnan(data)]
    # 绘制直方图
    sns.histplot(data, kde=True)
    plt.xlabel('value')
    plt.ylabel('frequency')
    plt.title(title)
    plt.show()
```

```python
cnt = 0
alpha = 0.05
for data1, data2, label in zip(dataB, dataR, header_list):
    data1 = data1[~np.isnan(data1)]
    data2 = data2[~np.isnan(data2)]
    # 前面6个选择t检验（因为有分布）
    # 后面两个选择Mann-Whitney U检验
    if cnt < 6:
        print("t_test:")
        t_statistic, p_value = ss.ttest_ind(data1, data2)
    else:
        print("u_test")
        u_statistic, p_value = ss.mannwhitneyu(data1, data2)
    print(label, ":")
    if p_value < alpha:
        print("两个数据不是来自同一个分布")
    else:
        print("两个数据来自同一个分布")
    print("__________________________________")
    cnt += 1
```
